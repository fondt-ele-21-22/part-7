## Esercizio: Random Walk Inerziale (1D)

**Consideriamo una variante del modello di Random Walk 1D**

Una particella si muove su _una retta_:

* La particella si muove per passi discreti ($+1$, $-1$)
* Ci sono 2 direzioni possibili: su, giù
* Ad ogni passo:
  - Con probabilità $p \in [0, 1]$ viene _mantenuta_ la direzione corrente
  - In caso contrario, la particella si muove nella _direzione opposta_

**Il mantenere la direzione può essere visto come un meccanismo di _inerzia_**

* Per $p = 1$ (i.e. $100\%$) la particella si muove in linea retta
* Per $p = 0$ (i.e. $0\%$) la particella oscilla
* Per $p = 0.5$ (i.e. $50\%$) otteniamo il modello di Random Walk 1D normale

## Esercizio: Random Walk Inerziale (1D)

**Lo stato della particella al passo $k$-mo**

...È rappresentato da _due quantità_

* La sua posizione $x_k$
* La sua direzione (una sorta di velocità) $v_k$

La velocità $v_k$ può valere $+1$ (su) o $-1$ (giù)

**L'evoluzione dello stato è definita dalla ricorsione:**

$$
(x_{k+1}, v_{k+1}) = \left\{\begin{aligned}
& (x_k+v_k, v_k) & \text{con probabilità $p$} \\
& (x_k-v_k, -v_k) & \text{altrimenti}
\end{aligned}\right.
$$
